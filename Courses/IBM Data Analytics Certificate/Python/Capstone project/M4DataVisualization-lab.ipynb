## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import matplotlib.pyplot as plt

conn=sqlite3.connect('m4_survey_data.sqlite')
query="SELECT ConvertedComp from master;"

df=pd.read_sql_query(query,conn)
conn.close()
plt.hist(df['ConvertedComp'], bins=10, color='skyblue', edgecolor='black')
plt.ticklabel_format(style='plain', axis='both')
plt.xticks(rotation=45)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of ConvertedComp')
plt.show()



```

### Box Plots


Plot a box plot of `Age.`


```python
conn=sqlite3.connect('m4_survey_data.sqlite')
query="SELECT Age from master;"

df=pd.read_sql_query(query,conn)
conn.close()
sns.boxplot(y=df['Age'],color='skyblue')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Your Column')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
conn=sqlite3.connect('m4_survey_data.sqlite')
query="SELECT Age,WorkWeekHrs from master;"

df=pd.read_sql_query(query,conn)
conn.close()

X=df['Age']
Y=df['WorkWeekHrs']

plt.scatter(X,Y)

plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.title('Scatter plot of Age and WorkWeekHrs')
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
conn=sqlite3.connect('m4_survey_data.sqlite')
query="SELECT Age,WorkWeekHrs,CodeRevHrs from master;"

df=pd.read_sql_query(query,conn)
conn.close()

X=df['WorkWeekHrs']
Y=df['CodeRevHrs']
BubbleSize=df['Age']

sns.scatterplot(data=df,x=X,y=Y,size=BubbleSize,sizes=(100,500),legend=False)

plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.title('Bubble plot of WorkWeekHrs and CodeRevHrs using Age as bubble size')

```

### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
conn=sqlite3.connect('m4_survey_data.sqlite')
query="SELECT Age,WorkWeekHrs,CodeRevHrs from master;"

df=pd.read_sql_query(query,conn)
conn.close()
age_filtered_df=df[(df['Age'] >= 30) & (df['Age'] <= 35)]
median_data=age_filtered_df[['WorkWeekHrs','CodeRevHrs']].median()

median_data.plot(kind='bar',stacked=True)
plt.title('Stacked chart of median WorkWeekHrs and CodeRevHrs for age group between 30-35')
plt.xticks(rotation=45)
plt.ylabel('Median Value')
plt.show()


```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
conn=sqlite3.connect('m4_survey_data.sqlite')
query="SELECT ConvertedComp, Age from master WHERE Age BETWEEN 45 AND 60;"

df=pd.read_sql_query(query,conn)
conn.close()



median_data=df.groupby('Age')['ConvertedComp'].median().reset_index()

plt.plot(median_data['Age'], median_data['ConvertedComp'], marker='o', linestyle='-')
plt.title('Median ConvertedComp by Ages 45-60')
plt.xlabel('Age')
plt.ylabel('ConvertedComp')
plt.show()


conn=sqlite3.connect('m4_survey_data.sqlite')
query="SELECT ConvertedComp, Age from master WHERE Age BETWEEN 25 AND 30;"

df=pd.read_sql_query(query,conn)
conn.close()



median_data=df.groupby('Age')['ConvertedComp'].median().reset_index()

plt.plot(median_data['Age'], median_data['ConvertedComp'], marker='o', linestyle='-')
plt.title('Median ConvertedComp by Ages 25-30')
plt.xlabel('Age')
plt.ylabel('ConvertedComp')
plt.show()
```
